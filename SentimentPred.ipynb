```
import numpy as np

# read data from text files
with open('/content/sample_data/reviews.txt', 'r') as f:
    reviews = f.read()
with open('/content/sample_data/label.txt', 'r') as f:
    labels = f.read()
```

```
print(type(reviews))
reviews = reviews.lower() # lowercase, standardize
all_text = ''.join([c for c in reviews if c not in punctuation])

# split by new lines and spaces
reviews_split = all_text.split('\n')
all_text = ' '.join(reviews_split)

# create a list of words
words = all_text.split()
```

```
words[:10]
```

```
from collections import Counter

## Build a dictionary that maps words to integers
counts = Counter(words)
print((counts.get))
vocab = sorted(counts, key=counts.get, reverse=True)
vocab_to_int = {word: ii for ii, word in enumerate(vocab, 1)}

## use the dict to tokenize each review in reviews_split
## store the tokenized reviews in reviews_ints
reviews_ints = []
for review in reviews_split:
    reviews_ints.append([vocab_to_int[word] for word in review.split()])
```

```
print('Unique words: ', len((vocab_to_int)))  # should ~ 74000+
print()

# print tokens in first review
print('Tokenized review: \n', reviews_ints[:2])
reviews_ints = reviews_ints[:25000][:]
```

```
# 1=positive, 0=negative label conversion
labels_split = labels.split('\n')
encoded = np.array([1 if label == 'positive' else 0 for label in labels_split])
```

```
review_lens = Counter([len(x) for x in reviews_ints])
print("Zero-length reviews: {}".format(review_lens[0]))
print("Maximum review length: {}".format(max(review_lens)))
```

```
print('Number of reviews before removing outliers: ', len(reviews_ints))

## remove any reviews/labels with zero length from the reviews_ints list.
l=[]

reviews_ints = [r if len(r)>0 else l.append(ii) for ii,r in enumerate(reviews_ints) ]
encoded_labels = []
for i in l:
  encoded.pop(i)
for e in encoded:
  encoded_labels.append(e) 
print(encoded_labels)
print(len(encoded)==len(encoded_labels))
encoded_labels = np.array(encoded_labels)
print(l)




print('Number of reviews after removing outliers: ', len(reviews_ints))
```

```
def pad_features(reviews_ints, seq_length):
    ''' Return features of review_ints, where each review is padded with 0's 
        or truncated to the input seq_length.
    '''
    ## implement function
    features = []
    for i,r in enumerate(reviews_ints):
      if(len(r)>seq_length):
        r = r[:seq_length]
        
      elif(len(r)<seq_length):
        shape_refined = seq_length-len(r)
        r = np.pad(r, (shape_refined,0), 'constant')
        #print(r)        

      len(reviews_ints[i])  

      features.append(r)


    
    features=np.array(features)
    
    return features
```

```
seq_length = 200

features = pad_features(reviews_ints, seq_length=seq_length)

## test statements - do not change - ##
assert len(features)==len(reviews_ints), "Your features should have as many rows as reviews."
print(len(features[0]))
assert len(features[0])==seq_length, "Each feature row should contain seq_length values."

# print first 10 values of the first 30 batches 
print(features[:30,:10])
```

```
split_frac = 0.8

## split data into training, validation, and test data (features and labels, x and y)

split_idx = int(len(features)*split_frac)
train_x, remaining_x = features[:split_idx], features[split_idx:]
train_y, remaining_y = encoded_labels[:split_idx], encoded_labels[split_idx:]

test_idx = int(len(remaining_x)*0.5)
val_x, test_x = remaining_x[:test_idx], remaining_x[test_idx:]
val_y, test_y = remaining_y[:test_idx], remaining_y[test_idx:]

## print out the shapes of your resultant feature data
print("\t\t\tFeature Shapes:")
print("Train set: \t\t{}".format(train_x.shape), 
      "\nValidation set: \t{}".format(val_x.shape),
      "\nTest set: \t\t{}".format(test_x.shape))
```

```
import torch
from torch.utils.data import TensorDataset, DataLoader

# create Tensor datasets
print(train_x.shape[0])
print(train_y.shape[0])

train_data = TensorDataset(torch.from_numpy(train_x), torch.from_numpy(train_y))
valid_data = TensorDataset(torch.from_numpy(val_x), torch.from_numpy(val_y))
test_data = TensorDataset(torch.from_numpy(test_x), torch.from_numpy(test_y))

# dataloaders
batch_size = 50

# make sure to SHUFFLE your data
train_loader = DataLoader(train_data, shuffle=True, batch_size=batch_size)
valid_loader = DataLoader(valid_data, shuffle=True, batch_size=batch_size)
test_loader = DataLoader(test_data, shuffle=True, batch_size=batch_size)
```
